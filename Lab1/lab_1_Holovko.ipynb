Відмінність інтерпретатора від компілятора:

Інтерпретатор виконує код безпосередньо, інтерпретуючи його рядок за рядком під час виконання програми. Він не створює проміжний файл і не перетворює програму в машинний код перед виконанням.
Компілятор переводить весь вихідний код програми в машинний код або байт-код, який може бути виконаний пізніше. Це відбувається на етапі компіляції, і компілятор створює окремий файл, який можна виконувати.

Python є інтерпретованою мовою програмування. Хоча Python використовує проміжний байт-код (що іноді створює ілюзію компіляції), в основному він працює через інтерпретатор, який виконуватиме байт-код під час роботи.


IDE PyCharm: PyCharm — це інтегроване середовище розробки (IDE) для програмування на Python. Воно пропонує різноманітні інструменти для зручної розробки: підсвічування синтаксису, автоматичне доповнення коду, інтеграція з системами контролю версій, налагодження (debugging), тестування, вбудовані утиліти для роботи з базами даних та багато інших можливостей.


Система контролю версій (СКВ): Система контролю версій — це програмне забезпечення, яке дозволяє зберігати різні версії файлів і відстежувати зміни в проєкті. Вона допомагає командам розробників працювати з кодом, дозволяючи відновлювати попередні версії, відстежувати зміни, а також зливати зміни з різних гілок проєкту.

 Відмінність SVN та Git:


SVN (Subversion) є централізованою системою контролю версій, де є один центральний репозиторій, і всі зміни відправляються та отримуються з цього репозиторію.
Git — це розподілена система контролю версій, де кожен розробник має локальну копію репозиторію, а зміни можна синхронізувати з віддаленим репозиторієм.
Що відбувається при виконанні команди git add .? Команда git add . додає всі зміни (створення, видалення чи зміна файлів) в поточному каталозі і його підкаталогах до індексу (staging area). Це означає, що файли підготовлені до коміту, але зміни ще не зафіксовані в історії репозиторію.


Що відбувається при виконанні команди git commit? Команда git commit створює новий коміт, який містить всі зміни, що були додані до індексу за допомогою git add. Коміт зберігає ці зміни в історії репозиторію. Кожен коміт має власний ідентифікатор і зазвичай містить опис змін, що були внесені.


Команда для розміщення локальної копії репозиторію на віддаленому сервері (GitHub): Для того, щоб розмістити локальний репозиторій на віддаленому сервері, наприклад, на GitHub, потрібно виконати команду:

 git push origin main
 Де origin — це стандартне ім'я для віддаленого репозиторію, а main — це ім'я гілки (може бути master в старих проєктах). Команда git push відправляє ваші локальні коміти на віддалений сервер.
